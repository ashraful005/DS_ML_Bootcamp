```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
x
```

```python
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot')
```

```python
y = x**3
```

```python
y
```

```python
plt.plot(x,y, 'y*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
```

```python
## Creating Subplots
plt.subplot(2,2,1) #(nrows, ncolms, index)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go--')
```

```python
x=np.arange(0, 4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine curve")
plt.plot(x,y,'r--')
```

# Bar plot

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
