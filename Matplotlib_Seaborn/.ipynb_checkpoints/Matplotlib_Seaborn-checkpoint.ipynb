```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
x
```

```python
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot')
```

```python
y = x**3
```

```python
y
```

```python
plt.plot(x,y, 'y*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
```

```python
## Creating Subplots
plt.subplot(2,2,1) #(nrows, ncolms, index)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go--')
```

```python
x=np.arange(0, 4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine curve")
plt.plot(x,y,'r--')
```

# Bar plot

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
a=np.array([22, 87, 5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes= [215,130,245,210]
colors= ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode=(0.4, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```

```python
# Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```

# Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="time")
```

```python
sns.distplot(df['tip'])
```

```python
# JoinPlot -> relationship of numerical features
sns.jointplot(x='tip', y='total_bill', data=df, kind='hist')
```

```python
# JoinPlot -> relationship of numerical features
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.countplot('sex', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```
